```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url="https://presidentofindia.nic.in/former-presidents.htm"
```

```python
page= requests.get(url)
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content,"html.parser")
print(soup.prettify)
```

```python
scraped_president = soup.find_all("h3")
scraped_president
```

```python
president_names =[]
for president in scraped_president:
    president= president.get_text().replace("(         )","")
    president_names.append(president)
president_names
```

```python
scraped_term = soup.find_all("p")
scraped_term
```

```python
terms = []
for term in scraped_term:
    term=term.get_text().strip()
    terms.append(term)
terms
terms= term.removeprefix("<p class="copyright">Copyright © 2017 The Rashtrapati Bhavan.</p>,
 <p clas_="lastUpdated"><span id_="ContentPlaceHolder1_UserPageLastUpdateDate1_lblPageLastUpdateDate">Page last updated on: <strong>16-September-2021 12:55 PM</strong></span>
 </p>,
 <p class_="govNic"><a href="http://www.nic.in/" target="_blank"><img alt="National Informatics Centre - www.nic.in - External website that opens in a new window" height="28" src="Images/nic-logo.gif" title="National Informatics Centre - www.nic.in - External website that opens in a new window" width="28"/></a>
  Website hosted by National Informatics Centre. Website Content provided by the President's Secretariat.</p>]")
terms
```

```python
df = pd.DataFrame({"Former President Name":president_names,"terms of Office":terms})
df
```
